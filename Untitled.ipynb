```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([0, 0, 0, 1, 1])

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def predict_prob(x, w, b):
    return sigmoid(w * x + b)

def predict(x, w, b):
    return predict_prob(x, w, b) >= 0.5

def loss(y_true, y_prob):
    return -np.mean(y_true * np.log(y_prob + 1e-15) + (1 - y_true) * np.log(1 - y_prob + 1e-15))

def train(x, y, lr=0.1, epochs=1000):
    w = 0.0
    b = 0.0

    for i in range(epochs):
        y_pred = predict_prob(x, w, b)
        dw = np.dot(x, (y_pred - y)) / len(x)
        db = np.mean(y_pred - y)

        w -= lr * dw
        b -= lr * db

        if i % 100 == 0:
            print(f"Epoch {i}: Loss = {loss(y, y_pred):.4f}")

    return w, b

w, b = train(x, y)

x_plot = np.linspace(0, 6, 100)
y_plot = predict_prob(x_plot, w, b)

plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x_plot, y_plot, color='red', label='Sigmoid Curve')
plt.xlabel("x")
plt.ylabel("Probability")
plt.title("Logistic Regression from Scratch")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Predict probability
def predict_prob(x, w, b):
    z = w * x + b
    return sigmoid(z)

# Binary cross-entropy loss
def loss(y, y_pred):
    return -np.mean(y * np.log(y_pred + 1e-10) + (1 - y) * np.log(1 - y_pred + 1e-10))

# Training function
def train(x, y, lr=0.1, epochs=10):
    w = 0.0  # initial weight
    b = 0.0  # initial bias

    for i in range(epochs):
        ### Step 1: Forward Pass ###
        z = w * x + b                           # Linear function
        y_pred = sigmoid(z)                     # Apply sigmoid

        ### Step 2: Compute Gradients ###
        error = y_pred - y                      # Difference between prediction and actual
        dw = np.dot(x, error) / len(x)          # Gradient of loss w.r.t weight
        db = np.mean(error)                     # Gradient of loss w.r.t bias

        ### Step 3: Update Weights ###
        w = w - lr * dw
        b = b - lr * db

        ### Step 4: Print Progress ###
        current_loss = loss(y, y_pred)
        print(f"Epoch {i+1}: Loss = {current_loss:.4f}, w = {w:.4f}, b = {b:.4f}")

    return w, b
w, b = train(x, y)    
x_plot = np.linspace(0, 6, 100)
y_plot = predict_prob(x_plot, w, b)

plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x_plot, y_plot, color='red', label='Sigmoid Curve')
plt.xlabel("x")
plt.ylabel("Probability")
plt.title("Logistic Regression from Scratch")
plt.legend()
plt.grid(True)
plt.show()
```
